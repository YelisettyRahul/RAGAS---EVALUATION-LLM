```python
import pandas as pd
import json

# Load the Excel file
file_path = "qca_dataset_specific_v2(Sheet1).csv"  # Update with your actual file path
df = pd.read_csv(file_path)

# Convert to JSON format
json_data = df.rename(columns={'question': 'question', 'answer': 'reference'}).to_dict(orient='records')

# Save JSON to a file or print
json_output_path = "output.json"
with open(json_output_path, "w", encoding="utf-8") as json_file:
    json.dump(json_data, json_file, indent=4, ensure_ascii=False)

# Print the JSON
print(json.dumps(json_data, indent=4, ensure_ascii=False))
```
